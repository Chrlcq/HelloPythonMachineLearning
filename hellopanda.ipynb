```python
import math
import random
import statistics
import os
import glob
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
#import seaborn as sns

print('----------pandas_titanic---------')

#lecture excel en dataframe
data = pd.read_excel('titanic3.xls')
#forme 
print(data.shape)
#entetes
print(data.columns)
#premieres lignes
print(data.head)
#supprimer colonnes
data = data.drop(['name','sibsp','parch','ticket','fare','cabin','embarked','boat','body','home.dest'],axis=1) 
print(data.columns)
print(data.head)
#statisques de base par colonne
print(data.describe())

#il y a des ages vides dans le dataset
#option 1 : remplacer les vides par la moyenne des ages
#data = data.fillna(data['age'].mean())
#option 2 : supprimer les lignes incompletes
data = data.dropna(axis=0)
# equivaut a : data.dropna(axis=0, inplace=true)
print(data.shape)#repartition par classe de voyageur
print(data['pclass'].value_counts())

#generation d'un tableau matplotlib
data['pclass'].value_counts().plot.bar()
plt.show()
plt.savefig("pandas1.png")
plt.clf()
data['pclass'].value_counts().plot.bar()
plt.savefig("pandas_titanic1.png")
plt.show()
plt.clf()

#histogramme des ages
data['age'].hist()
plt.savefig("pandas_titanic2.png")
plt.show()
plt.clf()
#group by
print(data.groupby(['sex']).mean())
print(data.groupby(['sex','pclass']).mean())

#vocabulaire :
#serie pandas :
#    tableau numpy 1d (1 colonne) de data 
#    + un index modifiable
#manipulation des series
#il est possible de modifier l'index
#ici, l'index passe de nombres aux noms
#data = data.set_index('name')
#print(data['age'])
#il est possible d'ajouter 2 series si elles partagent le meme index
#dataframe pandas : dictionnaire de series


#indexing
#10 premieres valeurs de age
print(data['age'][0:10])
#boolean indexing : mineurs uniquement
print(data[data['age']<18])
#indexing comme numpy : index localization
print(data.iloc[0:2,0:2])
#idem mais en travaillant avec une colonne
print(data.loc[0:2,['age','sex']])
#exercice : modifier l'age pour en faire 4 categories
data.loc[data['age']<20,'age'] = 0
data.loc[(data['age']>=20)&(data['age']<30),'age'] = 1
data.loc[(data['age']>=30)&(data['age']<40),'age'] = 2
data.loc[data['age']>=40,'age'] = 3
print(data.head)
#histogramme des ages
print(data['age'].value_counts())
plt.savefig("pandas_titanic3.png")
plt.show()
plt.clf()

#fonction map
#applique une fonction sur chaque donnee d'une colonne
data['age'].map(lambda x:x+1)
print(data['age'].value_counts())
#remplacer texte par valeurs
print(data['sex'].map({'male':0,'female':1}))
print(data['sex'].value_counts())
#autre methode
data['sex'].replace(['male','female'],[0,1])
#autre methode : pandas categorise automatiquement
data['sex'].astype('category').cat.codes
#fonction apply() : idem map mais pour tout le dataframe
```

```python
print('----------pandas_bitcoin---------')

#lecture csv
#passage d'un index numerique a un index datetime
#ici, la colone Date. on parse les dates
bitcoin=pd.read_csv('BTC-EUR.csv',index_col ='Date',parse_dates=True)
print(bitcoin.head())
#on peut preciser une periode
bitcoin['2019']['Close'].plot()
plt.savefig("pandas_bitcoin1.png")
plt.show()
plt.clf()

#fevrier 19
bitcoin['2019-02']['Close'].plot()
plt.savefig("pandas_bitcoin2.png")
plt.show()
plt.clf()

#de 2016 a 2019
bitcoin['2016':'2019']['Close'].plot()
plt.savefig("pandas_bitcoin3.png")
plt.show()
plt.clf()

#regroupement par frequence temporelle
#ici par mois
bitcoin['2019']['Close'].resample('M').plot()
plt.savefig("pandas_bitcoin4.png")
plt.show()
plt.clf()

#regroupement par frequence temporelle
#ici moyenne par mois
bitcoin['2019']['Close'].resample('M').mean().plot()
plt.savefig("pandas_bitcoin5.png")
plt.show()
plt.clf()

#regroupement par frequence temporelle
#ici equart type par 2 semaines
bitcoin['2019']['Close'].resample('2w').std().plot()
plt.savefig("pandas_bitcoin6.png")
plt.show()
plt.clf()

#graph multiple
bitcoin['2019']['Close'].plot()
bitcoin['2019']['Close'].resample('M').mean().plot(label='moyenne par mois',lw=3,ls=':',alpha=0.8)
bitcoin['2019']['Close'].resample('W').mean().plot(label='moyenne par semaine',lw=2,ls='--',alpha=0.8)
plt.legend()
plt.savefig("pandas_bitcoin7.png")
plt.show()
plt.clf()

#tableau contenant moyenne, std.. par semaine
m = bitcoin['2019']['Close'].resample('W').agg(['mean','std','min','max'])
plt.figure(figsize=(12,8))
m['mean']['2019'].plot(label='moyenne par semaine')
plt.fill_between(m.index, m['max'],m['min'],alpha=0.2, label = 'min max par semaine')
plt.legend()
plt.savefig("pandas_bitcoin8.png")
plt.show()
plt.clf()

#mouving average : on calcule la moyenne sur  7 jours, on se deplace d'un jour et on recommence
bitcoin['2019']['Close'].rolling(window =7, center = True).mean().plot(label='rolling avg')
#moyenne mobile exponentielle
bitcoin['2019']['Close'].ewm(alpha=0.2).mean().plot(label='rolling expo')
plt.legend()
plt.savefig("pandas_bitcoin9.png")
plt.show()
plt.clf()
```

```python
print('----------pandas_bitcoin_ethereum---------')

ethereum=pd.read_csv('ETH-EUR.csv',index_col ='Date',parse_dates=True)
plt.figure(figsize=(12,8))
ethereum['2019']['Close'].plot(label='ethereum')
bitcoin['2019']['Close'].plot(label='btc')
plt.legend()

plt.savefig("pandas_bitcoin_ethereum1.png")
plt.show()
plt.clf()

#regroupe les btc et eth dans le meme dataframe
btc_eth = pd.merge(bitcoin, ethereum, on='Date', how ='inner',suffixes=('_btc','_eth'))
btc_eth[['Close_btc','Close_eth']].plot(figsize=(12,8),subplots=True)
plt.legend()
plt.savefig("pandas_bitcoin_ethereum2.png")
plt.show()
plt.clf()

#coeff de correlation
correlations = btc_eth[['Close_btc','Close_eth']].corr()
print(correlations)

#import seaborn as sns
#sns.heatmap(correlations)
#sns.savefig("pandas_bitcoin_ethereum3.png")
#sns.clf()
```

```python
print('----------pandas_bitcoin_buy_sell---------')
#exercice : turtle strategy
#determiner s'il faut acheter ou vendre le btc
#en fonction des 28 dernier jours
#si valeur sup au max : acheter
#si val inf min : vendre
#on initialise des col avec 0
bitcoin['buy'] = np.zeros(len(bitcoin))
bitcoin['sell'] = np.zeros(len(bitcoin))
#on cacule le min / max sur periode
bitcoin['rollingmax'] = bitcoin['Close'].shift(1).rolling(window =28).max()
bitcoin['rollingmin'] = bitcoin['Close'].shift(1).rolling(window =28).min()
bitcoin.loc[bitcoin['rollingmax'] < bitcoin['Close'], 'buy'] =1
bitcoin.loc[bitcoin['rollingmin'] < bitcoin['Close'], 'sell'] =-1
#matplot poo
start = '2019'
end='2019'
fig, ax = plt.subplots(2, figsize=(12,8),sharex=True)
ax[0].plot(bitcoin['Close'][start:end])
ax[0].plot(bitcoin['rollingmin'][start:end])
ax[0].plot(bitcoin['rollingmax'][start:end])
ax[0].legend(['bitcoin','rollingmin','rollingmax'])
ax[1].plot(bitcoin['buy'][start:end])
ax[1].plot(bitcoin['sell'][start:end])
ax[1].legend(['buy','sell'])

#bitcoin['buy'][bitcoin['2019']['Close'].rolling(window =28).max()>bitcoin['2019']['Close']] = 1 
#bitcoin['2019']['buy'].plot()
#bitcoin['2019']['sell'].plot()
#plt.legend()
plt.savefig("pandas_bitcoin_buy_sell.png")
plt.show()
plt.clf()
```
