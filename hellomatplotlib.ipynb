```python
''' < block comment'''
import math
import random
import statistics
import os
import glob
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

print('--------traitement_image----------')
#import photo de lib scipy
from scipy import misc
#import PyQt4   
#import tkinter
#mpl.use('TkAgg')
#mpl.matplotlib_fname()
#mpl.get_backend()

#lecture et affichage avec matplotlib
#mpl.pyplot.show()
#plt.style.use('classic')
#tableau de pixel contenant image
face = misc.face(gray=True)
#enregistrement dans plt + passage en nuances de gris
plt.imshow(face, cmap=plt.cm.gray)
#affichage en console
#plt.show(block=True)
print(face.shape)
#plt.plot([1,2,3], [1,2,3])
plt.show(block=True)

#stocker dimensions avec shape
h=face.shape[0]
l=face.shape[1]
print(face.shape, h, l)
#slicing pour zoomer
zoomface=face[math.floor(0.25*h):math.floor(-0.25*h),math.floor(0.25*l):math.floor(-0.25*l) ]
plt.imshow(zoomface, cmap=plt.cm.gray)
plt.show(block=True)

#eclaicir les px les plus sombres
zoomface[zoomface > 150] = 250
plt.imshow(zoomface, cmap=plt.cm.gray)
#sauvegarde dans le repertoire courant
plt.savefig("graphique.png")
plt.show(block=True)


#purge de plt (pour une utilisation ulterieure)
plt.clf()
```

```python
print('--------pyplot_procedural----------')

x = np.linspace(0,2,10)
y = np.linspace(0,2,10)
y = x**2
#courbe
plt.plot(x,y)
plt.savefig("courbe1.png")
plt.show()
plt.clf()#options courbe
#label = nom
#lw = épaisseur du trait
#ls = style du trait 
#c = couleur du trait
plt.figure(figsize=(6,4))#optionnel taille en inch
plt.plot(x,y , c='red', lw=2,ls='--', label ='carre')
#ajout d'une courbe sur la meme figure
plt.plot(x,x**3,label='cube')
#nom axe
plt.xlabel('axe x')
plt.ylabel('axe y')
#titre
plt.title('ma figure')
#affiche les labels des courbes
plt.legend()
plt.savefig("courbe2.png" )
plt.show()

plt.clf()#nuage de points
plt.scatter(x,y)
plt.savefig("scatter1.png")
plt.show()
plt.clf()#afficher plusieurs figures en meme temps
#ex : 2 graphique sur 1 colonne. 
#Le dernier 1 signifie que l'on travaille dans le premier graph
plt.figure()
plt.subplot(2,1,1)
plt.plot(x,y,c='red')
plt.subplot(2,1,2)
plt.plot(x,np.cos(x))
plt.savefig("courbes1.png")
plt.show()
plt.clf()
```

```python
print('--------pyplot_poo----------')
#plus pousse. 
#sharex = permet de creer des graph qui pratage la meme absisse ou ordonnee
#creer objet figure et axe
fig, ax = plt.subplots(2,1,sharex=True)
ax[0].plot(x,y)
ax[1].plot(x,np.sin(x))
plt.savefig("courbe_poo1.png")
plt.show()
plt.clf()
```

```python
print('--------pyplot_scatter----------')
from sklearn.datasets import load_iris
iris = load_iris()
x=	iris.data
y= iris.target
names= list(iris.target_names)
#nuage de points (cf vid15 pour iris)
#alpha = transparence
#s = epaisseur du point (ex :100 ou var)
#c =couleur, ici 1 couleur pour chaque valeur de y (soit 3 coul) 
plt.scatter(x[:,0],x[:,1],c=y,alpha=0.5, s=100)
plt.xlabel('longueur sepal')
plt.ylabel('largeur sepal')
plt.savefig("scatter2.png")
plt.show()
plt.clf()
```

```python
print('--------pyplot_3d----------')
#graphiques 3d
from mpl_toolkits.mplot3d import Axes3D
ax = plt.axes(projection='3d')
ax.scatter(x[:,0],x[:,1],x[:,2],c=y )
plt.savefig("scatter3d.png")
plt.show()
plt.clf()

#modelisation
f = lambda x,y : np.sin(x) + np.cos(x+y)
X = np.linspace(0,5,100)
Y = np.linspace(0,5,100)
X,Y = np.meshgrid(X,Y)
Z = f(X,Y)
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap ='plasma')
plt.savefig("surface3d.png")
plt.show()
plt.clf()
```

```python
print('--------pyplot_histogramme----------')
x=	iris.data
y= iris.target#abscisses = longeur
#ordonnees = nb appartitions
#bins = nb sections (defaut 10)
plt.hist(x[:,0],bins=30)
#seconde valeur meme graph
plt.hist(x[:,1],bins=30)
plt.savefig("histogramme1.png")
plt.show()
plt.clf()


#en 2d
#bins = nb sections (defaut 10)
plt.hist2d(x[:,0],x[:,1],bins=30)
plt.xlabel('longueur sepal')
plt.ylabel('largeur sepal')
#legende par couleur
plt.colorbar()
plt.savefig("histogramme2d.png")
plt.show()
plt.clf()
#tableau de pixel contenant image
face = misc.face(gray=True)
#analyse de la repartition des niveaux de gris dans l'image plus haut
plt.hist(face.ravel(),bins=255)
plt.savefig("histogramme_image.png")
plt.show()
plt.clf()
```

```python
print('--------pyplot_contour_plot----------')#tranche sur le dessus
#comme une carte
#20 : niveau de detail
#si trait continu : val positives
#si trait discontinu : val negatives
plt.contour(X,Y,Z,20,colors='black')
plt.savefig("contour_plot1.png")
plt.show()
plt.clf()
plt.contour(X,Y,Z,20,cmap='RdGy')
plt.colorbar()
plt.savefig("contour_plot2.png")
plt.show()
plt.clf()
```
