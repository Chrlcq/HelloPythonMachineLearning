```python
print('--------manip_tab----------')#creation tab nombres aléatoires
q = np.random.randint(0,10,[2,3])
print(q)
#somme de tout le tableau
print(q.sum())
#somme par colonne (axe 0 = vertical)
print(q.sum(axis=0))
#somme par ligne (axe 1 = horizontal)
print(q.sum(axis=1))#somme cumulee par colonne (axe 0 = vertical)
print(q.cumsum(axis=0))#produit par colonne (axe 0 = vertical)
print(q.prod(axis=0))
print(q.cumprod(axis=0))
#min / max par colonne (axe 0 = vertical)
print(q.min(axis=0))
print(q.max(axis=0))#position (index) min / max par colonne (axe 0 = vertical)
print(q.argmin(axis=0))
print(q.argmax(axis=0))#tri de valeur ou les indexs par colonne (axe 0 = vertical)
print(q.sort(axis=0))
print(q.argsort(axis=0))#appliquer des fonctions math au tableau
#ici exponentielle
print(np.exp(q))#statistiques
#moyenne, variance et ecart type par colonne (axe 0 = vertical)
print(q.mean(axis=0))
print(q.var(axis=0))
print(q.std(axis=0))
#gestion des nan (valeur vide ou not a number)
print(np.nanmean(q))
print(np.nanvar(q))
print(np.nanstd(q))
#coef de correlation
print(np.corrcoef(q))
print(np.corrcoef(q)[0,1])#liste valeurs sans doublon (puis avec count par val)
print(np.unique(q))
print(np.unique(q,return_counts=True))
values, counts =np.unique(q,return_counts=True)
#tri en fonction du nombre d'apparitions
print(values[counts.argsort()])
#gestion des nan
#masque true si nan / false 
print(np.isnan(q))
#masque nombre de nan
print(np.isnan(q).sum())
#remplacer les nan par 0 dans un tableau 
q[np.isnan(q)]=0
```
