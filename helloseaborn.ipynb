```python
import math
import random
import statistics
import os
import glob
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd
import seaborn as sns

print('----------seaborn_iris---------')
from sklearn.datasets import load_iris
iris = load_iris()
x= iris.data
y= iris.target
names= list(iris.target_names)
print(names)
print(iris.feature_names)

#on transforme de tab numpy en dataframe
data = pd.DataFrame(data= np.c_[iris['data'], iris['target']], columns= iris['feature_names'] + ['variety'])
#print(data.head())
#toutes relations entre figures par variete
sns.pairplot(data,hue = 'variety')

plt.savefig("seaborn_iris1.png")
plt.show()
plt.clf()

```

```python
print('----------seaborn_titanic---------')
#dataset disponible dans sns
titanic = sns.load_dataset('titanic')
print(titanic.head)
#sns.pairplot(data)
#nuage de points
#repartition par age, classe et colore par sexe
sns.catplot(x='pclass',y='age',data=titanic,hue='sex')
plt.savefig("seaborn_titanic1.png")
plt.show()
plt.clf()

#boites a moustaches
sns.boxplot(x='pclass',y='age',data=titanic,hue='sex')
plt.savefig("seaborn_titanic2.png")
plt.show()
plt.clf()
#distribution
sns.distplot(titanic['fare'])
plt.savefig("seaborn_titanic3.png")
plt.show()
plt.clf()
#analyse 2 variables
sns.jointplot('age','fare',data=titanic)
plt.savefig("seaborn_titanic4.png")
plt.show()
plt.clf()
#analyse 2 variables
sns.jointplot('age','fare',data=titanic,kind='kde')
plt.savefig("seaborn_titanic5.png")
plt.show()
plt.clf()
#analyse 2 variables
sns.jointplot('age','fare',data=titanic,kind='hex')
plt.savefig("seaborn_titanic6.png")
plt.show()
plt.clf()
#correlation
sns.heatmap(titanic.corr())
plt.savefig("seaborn_titanic7.png")
plt.show()
plt.clf()
```
